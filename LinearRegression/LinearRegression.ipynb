## Implementing Simple Linear Regression from Scratch

Define a class `LR` to implement simple linear regression manually. The class contains methods for fitting the model to training data and making predictions on new data.


```python
class LR:
    def __init__(self):
        self.gradient=None
        self.y_intercept=None

    def fit(self,X_train,Y_train):
        print(X_train.shape)
        print(Y_train.shape)
        num=0
        den=0
        for i in range(X_train.shape[0]):
            num+=(X_train[i]-X_train.mean())*(Y_train[i]-Y_train.mean())
            den+=(X_train[i]-X_train.mean())**2
        self.gradient=num/den
        self.y_intercept=Y_train.mean()-(self.gradient*X_train.mean())
        print("The gradient is: ",self.gradient)
        print("the value of y intercept is: ",self.y_intercept)

    def predict(self,X_test):
        return self.gradient*X_test+self.y_intercept
```

## Splitting the Data

Use `train_test_split` to divide the CGPA and placement score data into training and testing sets, with 20% of the data reserved for testing. Set a random seed to ensure consistent results across runs.


```python
X_train,X_test,Y_train,Y_test=train_test_split(cgpa_data,placement_data,test_size=0.2,random_state=2)
print(X_train.shape)
print(Y_train.shape)
```

## Training and Visualizing the Custom Model

Create an instance of the custom linear regression class. Fit the model on the training data, then visualize both training and testing predictions using scatter plots for comparison.


```python
model=LR()

model.fit(X_train,Y_train)

y_pred=model.predict(X_train)
y_test_pred=model.predict(X_test)
plt.scatter(X_train,Y_train)
plt.scatter(X_train,y_pred)
plt.xlabel("Cgpa")
plt.ylabel("placement score")
plt.title("Training Data")
plt.show()

plt.scatter(X_test,Y_test)
plt.scatter(X_test,y_test_pred)
plt.xlabel("Cgpa")
plt.ylabel("placement score")
plt.title("Testing data")
plt.show()
```

## Custom Implementation of Multiple Linear Regression

Define a class to perform multiple linear regression using the Normal Equation. Add a bias column to the features and compute the weights using the pseudoinverse method.


```python
class MultipleLR:
    def __init__(self):
        self.gradient=None
        self.y_intercept=None
    def fit(self,X_train,Y_train):
        X_train = X_train.to_numpy().astype(float)
        Y_train = Y_train.to_numpy().astype(float)
        X_train=np.insert(X_train,0,1,axis=1).astype(float)
        Y_train=Y_train.astype(float)
        intercepts = np.linalg.pinv(X_train) @ Y_train

        # print(intercepts)
        self.y_intercept=intercepts[0]
        self.gradient=intercepts[1:]
        # print("...")
        # print(self.y_intercept)
        # print(self.gradient)
    def predict(self,X_test):
        return np.dot(X_test,self.gradient)+self.y_intercept
        
    
```
